```python
url_ = 'http://mashable.com/2016/03/04/the-who-50th-anniversary'
```

```python
from IPython.display import IFrame
IFrame(url_, 600, 600)
```

```python
import urllib.request
response = urllib.request.urlopen(url_)
html = response.read()
html[0:800]
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

print(soup.prettify())
```

```python
# Donde empieza y donde termina el texto

temp = str(soup)[str(soup).find('class="author_and_date"'):]

temp = temp[:temp.find('class="article-topics"')]
temp
```

```python
# Donde empieza y donde termina el texto

temp = str(soup)[str(soup).find('class="author_and_date"'):]

temp = temp[:temp.find('class="article-topics"')]
```

```python
imagenn = []
index = 0
while index < len(temp):
    index = temp.find('image', index)
    if index == -1:
        break
    imagenn.append(temp[index:index+100].split('"')[0])
    index += 2
imagenn
```

```python
len(imagenn)
```
